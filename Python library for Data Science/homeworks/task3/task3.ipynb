С помощью этого атрибута найдите сумму всех показателей важности, установите, какие два признака показывают наибольшую важность.

```python
import pandas as pd 
import numpy as np
from sklearn.datasets import fetch_california_housing
```

```python
dataset = fetch_california_housing()
dataset
```

```python
X = pd.DataFrame(dataset.data, columns=dataset.feature_names)
X
```

```python
y = pd.DataFrame(dataset.target, columns=dataset.target_names)
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42)
```

RandomForestRegressor используется в задачах регрессии, где требуется предсказать непрерывные значения (например, цены на жильё, температуру и т. д.).

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=1000, max_depth=12, random_state=42)
```

```python
y_train
```

```python
model.fit(X_train, y_train.values[:, 0])
```

```python
y_pred = model.predict(X_valid)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred, y_valid)
```

```python
y_pred_train = model.predict(X_train)
```

```python
r2_score(y_train, y_pred_train)
```
