## Приведите по 2 примера, когда лучше максимизировать Precision, а когда Recall.

### Precision

1) Производство товаров - брак или нет (брак не должен попадать на полки)
2) Будет плохая погода или нет (лучше посидеть дома и не расстроиться, чем расстроиться)

### Recall

1) Допуск студентов к экзамену - допущен или нет (нам нужно как можно меньше должников)
2) Заправлен ли транспорт до определенного урвоня или нет (лучше больше чем меньше)

## Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?

F-мера представляет собой гармоническое среднее между precision и recall. Она стремится к нулю, если любой из этих параметров стремится к нулю. Эту метрику мы хотим видеть высокой если нам важен баланс Precision и Recall.

Т.е. для F-меры важно чтобы обе метрики были как можно больше 0, для среднего арифметического же допускается "жертва" в пользу одной из двух. Точность и полнота должны способствовать равномерному росту показателя F1 в отличие от среднего арифметического.
