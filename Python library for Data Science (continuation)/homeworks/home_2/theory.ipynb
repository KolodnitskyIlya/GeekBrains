## В чём различие между зависимыми и независимыми выборками?

Зависимые выборки (или связанные выборки) предполагают, что каждому испытуемому одной выборки соответствует один и только один испытуемый из другой выборки. Это означает, что результаты двух выборок могут быть связаны, например, когда измеряются одни и те же объекты до и после какого-либо воздействия.

Независимые выборки не имеют взаимосвязи между собой. Это значит, что результаты одной выборки не влияют на результаты другой. Каждое наблюдение в одной выборке не связано с наблюдением в другой.

## Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Параметрические критерии

* Используются, когда данные распределены нормально или близки к нормальному распределению.
* Применяются, если известны параметры распределения (например, среднее и дисперсия).
* Подходят для больших выборок, где можно предположить, что выборка будет следовать нормальному закону распределения благодаря центральной предельной теореме.

Непараметрические критерии

* Используются, когда данные не распределены нормально или когда предположение о нормальности не может быть выполнено.
* Применяются для анализа данных с небольшими выборками, где сложно проверить нормальность распределения.
* Подходят для данных, измеренных на порядковой шкале или когда данные содержат много выбросов.

Параметрические критерии применяются для выборок с нормальным законом распределения. Формула расчета этих критериев содержат параметры выборки: среднее, дисперсии и др. Поэтому они называются параметрическими.

Пример: t-критерий Стьюдента


Непараметрические критерии - это группа статистических критериев, которые не включают в расчёт параметры вероятностного распределения и основаны на оперировании частотами или рангами.

Пример: критерий Пирсона (Хи-квадрат)

Для использования параметрических критериев необходимо чтобы распределение было известно с точностью до конечного числа параметров, а для непараметрических критериев не предполагается знание функционального вида распределения.
